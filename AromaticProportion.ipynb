#  **Load data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/data.csv')
df
```

# Data preparation

```python
Y = df["logS"]
X = df.drop("logS",axis = 1 )
```

```python
print(X.shape)
print(Y.shape)
```

# Data splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# MODEL Building AND EVALEUATION

## linear regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score , mean_squared_error
lr_train_results = r2_score(y_train,y_train_pred)
lr_test_results = r2_score(y_test,y_test_pred)
lr_train_mse = mean_squared_error(y_train,y_train_pred)
lr_test_mse = mean_squared_error(y_test,y_test_pred)
lr_train_r2 = r2_score(y_train,y_train_pred)
lr_test_r2 = r2_score(y_test,y_test_pred)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```
